The last 10 TEOAE signals are observed to be very similar (steady) for a person,
yet these signals are very different for different individuals --> i.e., intraclass variability is low, while interclass variability is high.

This could mean that different subject's TEOAE signals contain different frequency components.

So, we extract frequency information from the signals using Fast Fourier Transform (fft), power spectral density (psd) and autocorrelation --> these frequency related information is used as feature input to the machine learning algorithm.

```python
# This extracts 80 responses for each of the 54 subjects: 
# for each subject: a total of 80 last 10 responses corresponding to the 8 different recording conditions
#the 8 conditions are:
# for subject = i --> session = 0/1 , ear = left/right, buffer = A/B ( 2 sessions * 2 ears * 2 buffers)
import sklearn.preprocessing
import pandas
import numpy

allInputFeatsAllSubjects = pd.DataFrame([])
allOutputLabelsAllSubjects = pd.DataFrame([])

#temp=pd.DataFrame([])
for a in range(54):
    #for each subject a 
    df1 = pd.DataFrame([])
    #a=list(range(0,1))
    b=[0,1]              # sessions 0 and 1
    c=['left','right']   # ears 
    d=['A','B']          # buffers
    
    allFeatures =  pd.DataFrame([])
    allInputFeatures =  pd.DataFrame([])
    outputLabel =  pd.DataFrame([])
    fftValues =  pd.DataFrame([])
    psdValues=  pd.DataFrame([])
    autocorrValues = pd.DataFrame([])
        
    for q,r,s in [(q,r,s) for q in b for r in c for s in d]:
        
        # df1 contains all the 80 TEOAE signals of subject a
        df1=df1.append(subject0(a,q,r,s))
        
    # compute fft, power spectral density and autocorrelation values for the 80 TEOAE signals
    # taking only 1st 128 points of fft since most information lies here only
    
    df1.to_csv('steadyteoaeLast10.csv')
    
    fftVal=np.fft.fft(df1.iloc[:,5:],1024)
    fftVal= np.abs(fftVal)
    
    fftValues= pd.DataFrame(fftVal)
        
    freqs,psdVal=signal.welch(df1.iloc[:,5:])
    psdVal=np.abs(psdVal)
    psdValues=  pd.DataFrame(psdVal)
    
    df1t=df1.iloc[:,5:].transpose()
    autocorrValues = df1t.corrwith(df1t, axis = 0)
    
    #normalize fft and psd
    
    fftValues = fftValues.iloc[:,0:128]
    sklearn.preprocessing.normalize(fftValues, norm='l2', axis=1, copy=True, return_norm=False)
    sklearn.preprocessing.normalize(psdValues, norm='l2', axis=1, copy=True, return_norm=False)    
    
    #concatenate inout features; 
    #saving all input features and output labels of all 80 TEOAE SIGNALS x 54 subjects = 4320 rows
    allFeatures = pd.concat([fftValues.reset_index(drop=True),psdValues.reset_index(drop=True),autocorrValues.reset_index(drop=True)], axis=1)
    #allFeatures = pd.concat([fftValues,psdValues,autocorrValues],  ignore_index=True, sort=False)
    
    allInputFeatures = allFeatures.iloc[0:80,:] #don't know why after 80 rows random single column values are there
    #allInputFeatsAllSubjects = allInputFeatsAllSubjects.append(allFeatures.iloc[0:80,:])
    #allInputFeatsAllSubjects = pd.concat([allInputFeatsAllSubjects,allInputFeatures], axis=0)
    allInputFeatsAllSubjects = pd.concat([allInputFeatsAllSubjects.reset_index(drop=True),allInputFeatures.reset_index(drop=True)], axis=0)
    # the following gets you the class label <= subject no.[0,1,2,...,53]
    outputLabel = df1.iloc[:,0] 
    #allOutputLabelsAllSubjects = allOutputLabelsAllSubjects.append(outputLabel)
    #append didn't work for output so using pd.cat instead
    allOutputLabelsAllSubjects = pd.concat([allOutputLabelsAllSubjects.reset_index(drop=True),outputLabel.reset_index(drop=True)], axis=0)
    #print(allOutputLabelsAllSubjects.shape)
    
    ##############################################
    #to save inout features in separate file
    subject_id=df1['subject'].iloc[0]
        #session_id=df['session'].iloc[0]
        #ear_id=df['ear'].iloc[0]
        #buffer_id=df['buffer'].iloc[0]
    file_name="subject"+str(subject_id)+"_fft_psd_autocorr"+".csv"
    allInputFeatures.to_csv(file_name)
    ################################################
    
```

```python
X=allInputFeatsAllSubjects.iloc[:,130:256]
y=allOutputLabelsAllSubjects
y=y.astype(int)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```
